```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
# 導入LogisticRegression方法
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
%matplotlib inline

# 導入數據 
dataset = pd.read_csv(r'C:/Users/Cindy/Desktop/0728test.csv')
dataset
```

```python
X = dataset.iloc[:,1:24]
y = dataset['Output_A1']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
#定義分類器
knn = KNeighborsClassifier()
#進行分類
knn.fit(X_train, y_train.astype('int'))
KNeighborsClassifier()
y_predict=knn.predict(X_test)

y_test = np.array(list(y_test))
y_pred = np.array(y_predict)
#顯示實際值和預測值
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_predict.flatten()})
df
```

```python
# 計算準確率, 由于每次數據集劃分不同, 可能不一样
# fabs 絕對值
print( np.sum(np.fabs(y_predict - y_test)) / float(len(y_test))) 
```

```python
# 平均絕對誤差(MAE)、均方誤差(MSE)、均方根誤差(RMSE)
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_predict))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_predict))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_predict)))
```
