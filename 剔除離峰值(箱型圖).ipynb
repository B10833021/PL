```python
a=pd.DataFrame(Input_A1_001_024).describe()
a.loc['IQR'] ='' 
a.loc['fence_max'] ='' 
a.loc['fence_min'] ='' 
for i in range(24):
    a.iloc[8,i]=a.iloc[6,i]-a.iloc[4,i]
    a.iloc[9,i]=a.iloc[6,i] + 1.5 * a.iloc[8,i]
    a.iloc[10,i]=a.iloc[4,i] - 1.5 * a.iloc[8,i]
a
```
