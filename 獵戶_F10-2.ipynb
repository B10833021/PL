```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
```

```python
print('channelWidth:\n',pd.value_counts(f10['channelWidth'].values.flatten()))
print('BSSID:\n',pd.value_counts(f10['BSSID'].values.flatten()))
```

```python
bssid_1 = f10[f10['BSSID'] == 'da:61:62:06:8f:7c' ]
bssid_1
```

```python
#print('level:\n',pd.value_counts(bssid_1['level'].values.flatten()))
print('channelWidth:\n',pd.value_counts(bssid_1['channelWidth'].values.flatten()))
print('frequency:\n',pd.value_counts(bssid_1['frequency'].values.flatten()))
```

```python
# 设置图框的大小
fig = plt.figure(figsize = (10,6))
# 绘图
plt.plot(bssid_1['temp_loc'], # x轴数据
         bssid_1['level'], # y轴数据
         linestyle = '-', # 折线类型
         linewidth = 2, # 折线宽度
         color = 'steelblue', # 折线颜色
         marker = 'o', # 点的形状
         markersize = 2, # 点的大小
         markeredgecolor='black', # 点的边框色
         markerfacecolor='brown') # 点的填充色

plt.title('IB F10 BSSID = da:61:62:06:8f:7c ')
plt.xlabel('Level')
plt.ylabel('IB F10 Location')

# 剔除图框上边界和右边界的刻度
plt.tick_params(top = 'off', right = 'off')

# 为了避免x轴日期刻度标签的重叠，设置x轴刻度自动展现，并且45度倾斜
#fig.autofmt_xdate(rotation = 45)

# 显示图形
plt.show()
```
