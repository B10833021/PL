```python
import pandas as pd
# 導入train數據 
train = pd.read_csv(r'C:/Users/Cindy/Desktop/train.csv')
train
```

```python
test = pd.read_csv(r'C:/Users/Cindy/Desktop/test.csv')
test
```

```python
data = pd.concat([train,test])
data
```

```python
data = data.drop(labels=['Name','Ticket','Fare'],axis='columns')
data
```

```python
data.isnull().any()
```

```python
#缺失值以中位數代替
data['Age'] = data['Age'].fillna(data['Age'].median())
data
```

```python
#缺失值以no cabin代替
#取第一個字母
data['Cabin'] = data['Cabin'].fillna("No cabin")
data['Cabin'] = data['Cabin'].str[0:1]
data
```

```python
data.loc[data['Cabin'] =='N','Cabin']=0
data.loc[data['Cabin'] =='A','Cabin']=1
data.loc[data['Cabin'] =='B','Cabin']=2
data.loc[data['Cabin'] =='C','Cabin']=3
data.loc[data['Cabin'] =='D','Cabin']=4
data.loc[data['Cabin'] =='E','Cabin']=5
data.loc[data['Cabin'] =='F','Cabin']=6
data.loc[data['Cabin'] =='G','Cabin']=7
data.loc[data['Cabin'] =='T','Cabin']=8
data
```

```python
#缺失值以 S 代替
#將embarked的C改0，Q改1，S改2
data['Embarked'] = data['Embarked'].fillna("S")
data.loc[data['Embarked'] =='C','Embarked']=0
data.loc[data['Embarked'] =='Q','Embarked']=1
data.loc[data['Embarked'] =='S','Embarked']=2
data
```

```python
#Sex為male，則傳回0，為female，則傳回1
data.loc[data['Sex'] =='male','Sex']=0
data.loc[data['Sex'] =='female','Sex']=1
data
```

```python
#familynum家人數
#刪除parch,sibsp
data['familynum']=data['Parch']+data['SibSp']
data = data.drop(labels=['Parch','SibSp'],axis='columns')
data
```

```python
data.isnull().any()
```

```python
#查看有哪些不同的值，並計算每個值有多少個重複值
print('Survived:\n',pd.value_counts(data['Survived'].values.flatten()))
print('Pclass:\n',pd.value_counts(data['Pclass'].values.flatten()))
print('Sex:\n',pd.value_counts(data['Sex'].values.flatten()))
print('Cabin:\n',pd.value_counts(data['Cabin'].values.flatten()))
print('Embarked:\n',pd.value_counts(data['Embarked'].values.flatten()))
#survived 0:no,1:yes
#Pclass
#Sex 0:male,1:female
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split 
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
#建立一個模型model為線性迴歸模型
model = LinearRegression()
```

```python
X = train.drop(labels=['Survived','PassengerId'],axis='columns') #刪除這些欄位
y = train['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
#60%為Training set，40%為Test set
# test_size代表測試組比例。
# random_state代表設定隨機種子，讓測試結果可被重複
```

```python
#放進模型內訓練
model.fit(X_train,y_train)
```

```python
#predict 使用線性模型進行預測
y_pred = model.predict(X_test)
y_pred
```
