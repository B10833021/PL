```python
import pandas as pd
import os
import matplotlib.pyplot as plt 
import glob
```

```python
path =r'C:/Users/Cindy/Desktop/獵戶/IB_F10'
allFiles = glob.glob(path + "/*.csv")

list_ = []
length = []

for file_ in allFiles:
    df = pd.read_csv(file_,index_col=None, header=0)
    print(len(df))
    length.append(len(df))
    list_.append(df)

f10_ = pd.concat(list_, axis = 0, ignore_index = True)
```

```python
f10_
```

```python
f10_.insert(0,"temp_loc",0)
f10_
```

```python
cnt = 1
sma = 0
big = length[0]
for i in range(len(length)):
    while(cnt<=len(length)):
        f10_.iloc[sma:big,0]=cnt
        cnt+=1
        sma += length[cnt-2]
        big += length[cnt-1]
```
