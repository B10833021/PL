```python
import pandas as pd
csvdata = pd.read_csv("C:/Users/Cindy/Desktop/程式語言/期末報告/歷年各縣市垃圾清理統計表 (2).csv")#讀取資料
city_csvdata=csvdata[['縣市名稱','垃圾產生量(公噸)','Resource Recovery Rate','District']]#取出這三欄
total_csvdata=city_csvdata[:23]#取出資料前21行
total_csvdata
```

```python
import matplotlib.pyplot as plt
import geopandas as gp
import pandas as pd
import numpy as np
nep_districts = gp.read_file(r'C:/Users/Cindy/Downloads/L14-Taiwan Population/TWN_adm/TWN_adm2.shp',encoding='utf-8')
nep_districts = nep_districts[['NAME_2', 'geometry']]
nep_districts.rename(columns = {'NAME_2' : 'District'}, inplace = True)
nep_districts
```

```python
nep_districts.replace('Changhwa', 'Changhwa County', inplace = True)
nep_districts.replace('Chiayi', 'Chiayi County', inplace = True)
nep_districts.replace('Hsinchu', 'Hsinchu County', inplace = True)
nep_districts.replace('Hualien', 'Hualien County', inplace = True)
nep_districts.replace('Ilan', 'Ilan County', inplace = True)
nep_districts.replace('Keelung City', 'Keelung City', inplace = True)
nep_districts.replace('Miaoli', 'Miaoli County', inplace = True)
nep_districts.replace('Nantou', 'Nantou County', inplace = True)
nep_districts.replace('Penghu', 'Penghu County', inplace = True)
nep_districts.replace('Pingtung', 'Pingtung County', inplace = True)
nep_districts.replace('Taichung', 'Taichung', inplace = True)
nep_districts.replace('Tainan', 'Tainan', inplace = True)
nep_districts.replace('Taipei', 'New Taipei City', inplace = True)
nep_districts.replace('Taitung', 'Taitung County', inplace = True)
nep_districts.replace('Taoyuan', 'Taoyuan City', inplace = True)
nep_districts.replace('Yunlin', 'Yunlin County', inplace = True)
nep_districts
```

```python
nep_districts = nep_districts.merge(total_csvdata, on = 'District')
nep_districts = nep_districts.drop([1])
nep_districts
```

# 各縣市108年垃圾量之色塊圖

```python
from matplotlib.colors import Normalize
from matplotlib import cm
from pylab import *
ax=nep_districts.plot(column = '垃圾產生量(公噸)', cmap = 'PuBu', legend = False)
norm = Normalize(vmin=44176, vmax=1320616)
n_cmap = cm.ScalarMappable(norm=norm, cmap='PuBu')
ax.get_figure().colorbar(n_cmap)       # 設定漸層色圖之圖標
plt.xlabel("Longitude")                # 設定x軸標題及粗體
plt.ylabel("Latitude")                 # 設定y軸標題及粗體
plt.axis('equal')                      # 設定x軸上圖間距相等
plt.title("Total Waste in 108",fontsize = 13, fontweight = "bold", y = 1.1)   #設定圖之標題
plt.show()
```

# 台北市垃圾量

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mtick 

garbage = [1416762,1363063,1474287,1361756,1204019,1238396,895860,736943,638223,597557]
population = [2605374,2598493,2639939,2641312,2646474,2633802,2641856,2627138,2622472,2616375]
l = [i for i in range(10)]
lx = ['85','86','87','88','89','90','91','92','93','94'] #設定Ｘ軸的刻度

fig = plt.figure()  
ax1 = fig.add_subplot(111) #表示1x1網格，第一個子圖 
ax1.plot(l, garbage,'or-',linewidth = 4,ms = 10, label=u'Garbage');#設定折線圖的線條粗細、點的大小、標籤
ax1.legend(loc=1) #標籤in upper right
ax1.set_ylim([400000,1700000 ]);#設定Ｙ軸（垃圾量）取值範圍
ax1.set_ylabel('Number');
plt.legend(loc="upper left") #標籤放在左上方

ax2 = ax1.twinx() #和ax1共用X軸
plt.bar(l,population,alpha=0.5,color='pink',label=u'Population')#設定長條圖的顏色、透明度、 標籤 
ax2.legend(loc=2) #標籤in upper left
ax2.set_ylim([2500000, 2800000])#設定Ｙ軸（人口）取值範圍

plt.title("Garbage in Taipei")
plt.legend(loc="upper right")#標籤放在右上方
plt.xticks(l,lx)
plt.style.use("seaborn-white") #設定圖表的樣式
plt.rcParams['savefig.dpi'] = 300 
plt.rcParams['figure.dpi'] = 300 #設定畫質
plt.savefig('garbagetpe.png') #存檔
plt.show()
```
